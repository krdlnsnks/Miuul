```python
bins = [-1, 7, 30, 90, df["SaleCheckInDayDiff"].max()]
labels = ["Last Minuters", "Potential Planners", "Planners", "Early Bookers"] #SaleCheckInDayDiff değişkenini kategorik bir değişkene çeviriniz
```

```python
df["EB_Score"] = pd.cut(df["SaleCheckInDayDiff"], bins, labels=labels)
df.head(50).to_excel("eb_scorew.xlsx", index=False)
```

```python
df.groupby(by=["SaleCityName", 'ConceptName', "EB_Score" ]).agg({"Price": ["mean", "count"]}) #COUNTRY, SOURCE, SEX, AGE kırılımında ortalama kazançlar nedir?
```

```python
df.groupby(by=["SaleCityName", "ConceptName", "Seasons"]).agg({"Price": ["mean", "count"]})
```

```python
df.groupby(by=["SaleCityName", "ConceptName", "CInDay"]).agg({"Price": ["mean", "count"]})
```

```python
agg_df = df.groupby(["SaleCityName", "ConceptName", "Seasons"]).agg({"Price": "mean"}).sort_values("Price", ascending=False)
agg_df.head(20) #City-Concept-Season kırılımın çıktısını PRICE'a göre sıralayınız
```

```python
agg_df.reset_index(inplace=True)
agg_df.head() #Indekste yer alan isimleri değişken ismine çeviriniz
```

```python
agg_df['sales_level_based'] = agg_df[["SaleCityName", "ConceptName", "Seasons"]].agg(lambda x: '_'.join(x).upper(), axis=1) #Yeni level based satışları tanımlayınız ve veri setine değişken olarak ekle
```

```python
agg_df["SEGMENT"] = pd.qcut(agg_df["Price"], 4, labels=["D", "C", "B", "A"])
agg_df.head(30)
agg_df.groupby("SEGMENT").agg({"Price": ["mean", "max", "sum"]})

#Personaları segmentlere ayırma
```
